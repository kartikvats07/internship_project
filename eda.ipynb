```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv('cleaned_BTech_Survey.csv')
sns.set(style='whitegrid')  # Set visual style

# -------------------------------
# 🔢 1. Descriptive Stats
# -------------------------------
print("Descriptive Stats for CGPA and Study Hours:")
print("-" * 50)
for col in ['Current CGPA (out of 10)', 'Average Study Hours per Week']:
    print(f"\n{col}:")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Mode: {df[col].mode()[0]:.2f}")

# -------------------------------
# 📚 2. Group Averages
# -------------------------------
print("\nBranch-wise CGPA Averages:")
print(df.groupby('Branch')['Current CGPA (out of 10)'].mean().sort_values(ascending=False))

print("\nYear-wise CGPA Averages:")
print(df.groupby('Year of Study')['Current CGPA (out of 10)'].mean().sort_values())

# -------------------------------
# 📊 3. Participation Metrics
# -------------------------------
print("\nInternship Participation (%):")
print(df['Have you done any internships?'].value_counts(normalize=True) * 100)

print("\nAverage Number of Clubs Participated In:")
print(df['Number of Technical Clubs or Competitions Participated In'].mean())

# -------------------------------
# 📈 4. Histograms: CGPA & Study Hours
# -------------------------------
plt.figure(figsize=(7, 4))
sns.histplot(df['Current CGPA (out of 10)'], bins=10, kde=True, color='skyblue')
plt.title("Histogram: CGPA Distribution")
plt.xlabel("CGPA")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 4))
sns.histplot(df['Average Study Hours per Week'], bins=10, kde=True, color='lightgreen')
plt.title("Histogram: Study Hours per Week")
plt.xlabel("Hours/Week")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# -------------------------------
# 📊 5. Bar Chart: Career Preferences
# -------------------------------
plt.figure(figsize=(9, 5))
sns.countplot(y='Preferred Career choice', data=df, order=df['Preferred Career choice'].value_counts().index, palette='muted')
plt.title("Bar Chart: Preferred Career Choices")
plt.xlabel("Number of Students")
plt.ylabel("Career Option")
plt.tight_layout()
plt.show()

# -------------------------------
# 🥧 6. Pie Chart: Branch-wise Distribution
# -------------------------------
branch_counts = df['Branch'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(branch_counts, labels=branch_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Pie Chart: Branch-wise Distribution")
plt.tight_layout()
plt.show()

# -------------------------------
# 📊 7. Boxplot: CGPA vs Internship
# -------------------------------
plt.figure(figsize=(7, 4))
sns.boxplot(x='Have you done any internships?', y='Current CGPA (out of 10)', data=df, palette='cool')
plt.title("CGPA vs Internship Experience")
plt.xlabel("Internship Done")
plt.ylabel("CGPA")
plt.tight_layout()
plt.show()

# -------------------------------
# 📊 8. Challenges Faced Countplot
# -------------------------------
plt.figure(figsize=(9, 5))
sns.countplot(y='Main Challenges Faced During Study', data=df, order=df['Main Challenges Faced During Study'].value_counts().index, palette='Set2')
plt.title("Main Academic Challenges Faced by Students")
plt.xlabel("Number of Students")
plt.tight_layout()
plt.show()

# -------------------------------
# 📊 9. Suggestions for Support
# -------------------------------
plt.figure(figsize=(9, 5))
sns.countplot(y='Any Suggestions to Improve Academic or Career Support', data=df, order=df['Any Suggestions to Improve Academic or Career Support'].value_counts().index, palette='Set3')
plt.title("Student Suggestions for Academic or Career Support")
plt.xlabel("Number of Students")
plt.tight_layout()
plt.show()
```
